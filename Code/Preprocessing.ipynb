```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./Data/201904_link_01.txt')
```

```python
df
```

```python
df = df[['od_id', 'index', 'seq', 'lon', 'lat',
         'sum_len', 'track_in_date', 'track_in_time', 'track_sum_travel_time']]
```

```python
first_index = df[(df['index']== 0) & (df['seq'] == 0)].index
first_index
```

```python
for i in range (len(first_index)):
    id_in = df.loc[first_index[i], 'od_id']
    if(i != len(first_index)-1):
        file_out = df.loc[first_index[i]:(first_index[i+1]-1)]
        file_out.to_csv('./id_based_parsing/id_{}.csv'.format(id_in))
```

```python
##Summary

df = pd.read_csv('./Data/4_summary.txt')
```

```python
df
```

```python
IDs = df['od_id'].values
print(len(IDs))
IDs
```

```python
for i, ids in enumerate(IDs):
    df1 = pd.read_csv('./Data/id_based_parsing/id_{}.csv'.format(ids))
    df2 = df1.loc[0]
    df2['lon'] = df1['lon'].values
    df2['lat'] = df1['lat'].values
    tot_time = df1.loc[df1.index[-1], 'track_sum_travel_time']
    tot_dist = df1.loc[df1.index[-1], 'sum_len']
    
    ##make time from data
    df2['time'] = tot_time / 10
    df2['dist'] = tot_dist / 10000
    time = df1.loc[0, 'track_in_time']
    date = df1.loc[0, "track_in_date"]
    time = str(date) + str('%6d' %time)
    
    time = time.replace(' ', '0')  ## 20190425211320
    time = pd.to_datetime(time) # 2019-04-25 21:13:20
    weekID = time.weekday() # 3
    
    df2['weekID'] = weekID
    df2['dateID'] = time.day
    timeID = (time.hour * 60) + (time.minute) + (time.second / 60) # 1273.333
    df2['timeID'] = timeID

    df2 = df2.rename({'od_id': 'driverID'})
    df2 = df2.rename({'lon': 'lngs'})
    df2 = df2.rename({'lat': 'lats'})
    driver =df2['driverID']
    driver = driver.replace('B', '')
    driver = int(driver)
    df2['driverID'] = driver
    df2 = df2.drop(['Unnamed: 0', 'index', 'point_in_time','seq', 'flink', 'tlink', 
                    'turn', 'flink_len', 'sum_len', 'facil', 'road_type', 'link_type',
                    'light', 'lane', 'outlink_cnt', 'zone', 'track_in_stamp', 
                    'track_in_date', 'track_in_time', 'track_speed', 'track_travel_time',
                    'track_sum_travel_time', 'static_spd', 'hq_speed', 'ol_rot_spd', 
                    'ol_spd', 'svc_name'])
    states = [1]*len(df2['lngs'])
    time_gap = [1]*len(df2['lngs'])
    dist_gap = [1]*len(df2['lngs'])
    df2['states']=states
    df2['time_gap']=time_gap
    df2['dist_gap']=dist_gap
    # to json
    df2.to_json('./Data/Test_data/{}.txt'.format(i, '06'))    
```
